```python
import pandas as pd
```

```python
#create a list of products
fruits = ['apple', 'banana', 'mangoe', 'pineapple', 'orange', 'pear', 'tomatoe']
```

```python
#convert list to pd series
fruits_series = pd.Series(fruits)
```

```python
#create a series with prices
prices = [70, 10, 30, 80, 10, 30, 50]
prices_series = pd.Series(prices)
```

```python
#add the two series together
fruits_prices = fruits_series + ' costs KES ' + prices_series.astype(str)
print(fruits_prices)
```

```python
#create a df from the dict of lists
fruits_prices_dict = {'fruits': fruits, 'prices': prices}
fruits_prices_df = pd.DataFrame(fruits_prices_dict)
print(fruits_prices_df)
```

```python
#create a df from a list of dicts
fruits_prices_list = [{'fruits': fruits[i], 'prices': prices[i]} for i in range(len(fruits))]
fruits_prices_df2 = pd.DataFrame(fruits_prices_list)
print(fruits_prices_df2)
```
